```python
import pandas as pd
```

```python
data=pd.read_csv('/content/titanic_clean.csv')
```

```python
summary=data.describe()
```

```python
print (summary)
```

```python
print (data.describe(include='object'))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['age'].dropna(), bins=20,edgecolor='black',color='skyblue')
plt.title('Distribution of age')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['pclass'].value_counts().plot(kind='bar',color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
```

```python

sns.pairplot(data[['age','fare','sibsp','parch']].dropna())
plt.show()
```

```python
correlations_matrix=data[['age','fare','sibsp','parch']].corr()
sns.heatmap(correlations_matrix,annot=True,cmap='coolwarm',cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='survived',data=data,palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='pclass',y='age',data=data,palette='muted')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```
