```python
import pandas as pd
```

```python
data=pd.read_csv('/content/titanic (1).csv')
```

```python
print (data.isnull().sum())
```

```python
data.head()
```

```python
data.fillna({'Age':data['Age'].mean()},inplace=True)
```

```python
data.fillna({'Age':data['Age'].median()},inplace=True)
```

```python
data.fillna({'Embarked':data['Embarked'].mode()},inplace=True)
```

```python
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scalar=MinMaxScaler()
```

```python
columns_to_normalize=['Age','Fare']
```

```python
data[columns_to_normalize]=Scalar.fit_transform(data[columns_to_normalize])
```

```python
data[columns_to_normalize].head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
data[columns_to_normalize]=scaler.fit_transform(data[columns_to_normalize])
```

```python
data[columns_to_normalize].head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['Age'],bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
plt.scatter(data['Age'],data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```
